```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
x , sr_x = librosa.load('E:/프로젝트 데이터/output_music/01 Kingstar.wav')
plt.figure(figsize=(14,5))
plt.title('10cm')
librosa.display.waveplot(x[:22000*15], sr=sr_x)
plt.savefig('./figures/sam_fig.png',dpi = 300)
```
